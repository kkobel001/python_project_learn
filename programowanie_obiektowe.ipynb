```python
product_dict = {"name": "Laptop", "net_price": 3200.0, "vat_rate": 0.23}
def gross_price(net_price, vat_rate):
    return net_price * (1 + vat_rate)

print("gross(dict_function):", gross_price(product_dict["net_price"],product_dict["vat_rate"]))

#instrukcja
class Product:
  #zawsze jest init self. obiekt ma swoje wlasne dane
  def __init__(self, name, net_price,vat_rate):
    self.name=name
    self.net_price=net_price
    self.vat_rate=vat_rate

  def gross_price(self):
    return self.net_price * (1+self.vat_rate)

#Obiekt
laptop=Product("Laptop", 3200.0, 0.23)
phone=Product("Phone", 1500.0, 0.18)

print("gross(class):", laptop.gross_price())
print("gross(class):", phone.gross_price())

#kazdy obiekt stworzony z tej samej klasy ma takiego same metody, ale różne wartosci w
```
