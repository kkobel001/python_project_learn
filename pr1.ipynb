```python
products=["laptop", "phone", "headphones"]

laptop_data=(1,"laptop", 3200.00)
phone_data=(2,"phone", 2100.00)
headphones_data=(2,"headphones", 550.00)

# print(type(laptop_data))

sales = {
    "laptop": 15
    ,"phone": 25
    ,"headphones":40
  }

print("PRODUCTS")
print(products)


total_uinit= sum(sales.values())
print(f"total units sold : {total_uinit}")

laptop_revenue= sales["laptop"]*laptop_data[2]
print(f"Total revenue za laptoppy to : {laptop_revenue}")


phone_revenue= sales["phone"]*phone_data[2]
print(f"Total revenue za phone to : {phone_revenue}")

headphones_revenue= sales["headphones"]*headphones_data[2]
print(f"Total revenue za headphones to : {headphones_revenue}")

total_revenue=laptop_revenue+phone_revenue+headphones_revenue
print(f"Suma sold revenue {total_revenue}")
```
