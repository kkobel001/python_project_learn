```python
import pandas as pd

data={
    "product" : ["Coffe", "Tea", "Water","Tea", "Water","Tea", "Water"],
    "price" : [12,8,3,12,8,3,4],
    "units":[120,80,200,120,80,200,100],
    "region":["North","North","North","South","South","South","West"]
}

sales=pd.DataFrame(data)
print(sales)

print(sales.groupby("product")["units"].sum())

print(sales.groupby(["product","region"])["units"].agg(["sum","mean"]))
```

```python
# import pandas as pd

#join


products=pd.DataFrame({
     "product" : ["Coffe", "Tea", "Water","Juice"],
     "category": ["Hot drink","Hot drink","Cold drink","Cold drink",]
})

sales_summary=pd.DataFrame({
     "product" : ["Coffe", "Tea", "Water"],
     "total units" : [270,180,450]
})

print(products)
print(sales_summary)


#join funkcja merged on=klucz wspolny


marged=pd.merge(products,sales_summary, on="product",how="left")
print(marged)

#union
sales_q1 = sales_summary.copy()
sales_q1["quarter"] ="Q1"
sales_q2 = sales_summary.copy()
sales_q2["quarter"] ="Q2"

concat_df= pd.concat([sales_q1,sales_q2])
print(concat_df)
```
