```python
import pandas as pd

data={
    "product" : ["Coffe", "Tea", "Water"],
    "price" : [12,8,3],
    "units":[120,80,200]
}

df=pd.DataFrame(data)
print(df.loc[0:1,["price", "units"] ])

#loc - jelsi chcemy się dostac do jakiegos dicta.. zawsze wyszukuje po indeksie , wiec jak chcemy 2 rekordy,
#cd .. to wyswietli nam ten który jest pod 0 i 1, wiec wpisujemy 0:1
#iloc - jesli chcemy dostac sie do dicta, ale wyszukuje po kolejnosci, a nie indeksie jak loc , a
# ilosc- jak chcemy wyswietlic 2 rekordy to tez wyswietli 0 i 1, ale wpisujemy 0:2


# nir dziala, bo nie dziala, bo iloc nie dziala na nazwach kolumn. iloc dziala na kolejnosci
# print(df.iloc[0:2,["price", "units"]])
#poprawnie:
print(df.iloc[0:2,1:3])


#maski boolowe /. jednowymiarowa tablica
df["units"] > 100

#jak chcemy pokazac , które dokladne rekordy sa powyzej 100
print(df.loc[df["units"] > 100, ["product","units"]])
```

```python
data={
    "product" : ["Coffe", "Tea", "Water","Juice"],
    "price" : [12,8,None,3],
    "units":[120,None,200,80],
    "region":["North","South","East", None]
}

df=pd.DataFrame(data)
df


# print(df.isna())
print(df.isna().any())

print(df.describe())


#wartosci ktore nam brakuje
#1 podejscie-usuwamy te ktore sa puste
print(df.dropna())
# ta metode stosuje sie jedynie wtedy kiedy jestesmy pewni ze chcemy pracowac jedynie z danymi pelnymi, poniewaz on usuwa cale rekordy

#brak unitow = zero:
df_filled=df.fillna({"price": 10, "units":0, "region": "Unknown"})
print(df_filled)
```

```python
df_filled.sum()
df_filled.max()
print(df_filled[["price", "units"]].min())

#na poziomie kolumny
print(df_filled[["price", "units"]].max())
#na poziomie rekordu
print(df_filled[["price", "units"]].max(axis=1))


```
