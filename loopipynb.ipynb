```python
sales = { "laptop": 15,"phone": 25,"headphones":40}

#znajdz pierwszy produkt, którego sprzedaż przekracza 20 sztuk
for pair in sales.items():
  key,values=pair
  if values > 20:
    print(key)
    break
  print("in for")
print("outside of for")


#znajdz pierwszy produkt, którego ponizej  30 sztuk
for pair in sales.items():
  key,values=pair
  if values <= 30:
    print(key)
    continue
  print("in for")
print("outside of for")
```
